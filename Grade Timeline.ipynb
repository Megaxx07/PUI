```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df16 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/Copy of 2016_latest_record.csv')
df17 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/Copy of 2017_latest_record.csv')
df18 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/Copy of 2018_latest_record.csv')
df19 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/Copy of 2019_latest_record.csv')
df20 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/Copy of 2020_latest_record.csv')
df21 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/Copy of 2021_latest_record.csv')
```

```python
df16.head()
```

```python

A16 = df16.groupby(['CUISIN','INI_GRADE']).agg({'CAMIS':'count'}).reset_index()
A16all = df16.groupby(['CUISIN']).agg({'CAMIS':'count'}).reset_index()
A16 = df16.groupby(['CUISIN','INI_GRADE']).agg({'CAMIS':'count'}).reset_index()
A16 = A16.loc[A16['INI_GRADE'].isin(['A'])]
A16 = A16.rename(columns={'CAMIS':'COUNT'})
A16.head()
```

```python
A16['PERCENT'] = A16.apply(lambda x: x['COUNT'] / 
                                                 A16all.loc[A16all['CUISIN']==x['CUISIN']]['CAMIS'].values[0],axis = 1)
A16 = A16.rename(columns = {'PERCENT':'16percent'})
A16 = A16.drop(columns = {'INI_GRADE','COUNT'})
A16
```

```python
A17 = df17.groupby(['CUISIN','INI_GRADE']).agg({'CAMIS':'count'}).reset_index()
A17all = df17.groupby(['CUISIN']).agg({'CAMIS':'count'}).reset_index()
A17 = df17.groupby(['CUISIN','INI_GRADE']).agg({'CAMIS':'count'}).reset_index()
A17 = A17.loc[A17['INI_GRADE'].isin(['A'])]
A17 = A17.rename(columns={'CAMIS':'COUNT'})
A17['PERCENT'] = A17.apply(lambda x: x['COUNT'] / 
                                                 A17all.loc[A17all['CUISIN']==x['CUISIN']]['CAMIS'].values[0],axis = 1)
A17 = A17.rename(columns = {'PERCENT':'17percent'})
A17 = A17.drop(columns = {'INI_GRADE','COUNT'})
A17
```

```python
A18 = df18.groupby(['CUISIN','INI_GRADE']).agg({'CAMIS':'count'}).reset_index()
A18all = df18.groupby(['CUISIN']).agg({'CAMIS':'count'}).reset_index()
A18 = df18.groupby(['CUISIN','INI_GRADE']).agg({'CAMIS':'count'}).reset_index()
A18 = A18.loc[A18['INI_GRADE'].isin(['A'])]
A18 = A18.rename(columns={'CAMIS':'COUNT'})
A18['PERCENT'] = A18.apply(lambda x: x['COUNT'] / 
                                                 A18all.loc[A18all['CUISIN']==x['CUISIN']]['CAMIS'].values[0],axis = 1)
A18 = A18.rename(columns = {'PERCENT':'18percent'})
A18 = A18.drop(columns = {'INI_GRADE','COUNT'})
A18
```

```python

A19 = df19.groupby(['CUISIN','INI_GRADE']).agg({'CAMIS':'count'}).reset_index()
A19all = df19.groupby(['CUISIN']).agg({'CAMIS':'count'}).reset_index()
A19 = df19.groupby(['CUISIN','INI_GRADE']).agg({'CAMIS':'count'}).reset_index()
A19 = A19.loc[A19['INI_GRADE'].isin(['A'])]
A19 = A19.rename(columns={'CAMIS':'COUNT'})
A19['PERCENT'] = A19.apply(lambda x: x['COUNT'] / 
                                                 A19all.loc[A19all['CUISIN']==x['CUISIN']]['CAMIS'].values[0],axis = 1)
A19 = A19.rename(columns = {'PERCENT':'19percent'})
A19 = A19.drop(columns = {'INI_GRADE','COUNT'})
A19
```

```python
A20 = df20.groupby(['CUISIN','INI_GRADE']).agg({'CAMIS':'count'}).reset_index()
A20all = df20.groupby(['CUISIN']).agg({'CAMIS':'count'}).reset_index()
A20 = df20.groupby(['CUISIN','INI_GRADE']).agg({'CAMIS':'count'}).reset_index()
A20 = A20.loc[A20['INI_GRADE'].isin(['A'])]
A20 = A20.rename(columns={'CAMIS':'COUNT'})
A20['PERCENT'] = A20.apply(lambda x: x['COUNT'] / 
                                                 A20all.loc[A20all['CUISIN']==x['CUISIN']]['CAMIS'].values[0],axis = 1)
A20 = A20.rename(columns = {'PERCENT':'20percent'})
A20 = A20.drop(columns = {'INI_GRADE','COUNT'})
A20
```

```python
A21 = df21.groupby(['CUISIN','INI_GRADE']).agg({'CAMIS':'count'}).reset_index()
A21all = df21.groupby(['CUISIN']).agg({'CAMIS':'count'}).reset_index()
A21 = df21.groupby(['CUISIN','INI_GRADE']).agg({'CAMIS':'count'}).reset_index()
A21 = A21.loc[A21['INI_GRADE'].isin(['A'])]
A21 = A21.rename(columns={'CAMIS':'COUNT'})
A21['PERCENT'] = A21.apply(lambda x: x['COUNT'] / 
                                                 A21all.loc[A21all['CUISIN']==x['CUISIN']]['CAMIS'].values[0],axis = 1)
A21 = A21.rename(columns = {'PERCENT':'21percent'})
A21 = A21.drop(columns = {'INI_GRADE','COUNT'})
A21
```

```python
TOTAL = pd.merge(A16,A17, on='CUISIN',how='inner')
TOTAL = pd.merge(TOTAL,A18, on='CUISIN',how='inner')
TOTAL = pd.merge(TOTAL,A19, on='CUISIN',how='inner')
TOTAL = pd.merge(TOTAL,A20, on='CUISIN',how='inner')
TOTAL = pd.merge(TOTAL,A21, on='CUISIN',how='inner')

TOTALt5 = TOTAL.loc[TOTAL['CUISIN'].isin(['American','Chinese','Italian','Japanese','Mexican'])]
TOTALt5 = TOTALt5.set_index('CUISIN')
TOTALt5 = TOTALt5.T
TOTALt5
```

```python

plt.figure(figsize=(13,6))
x_axis = ['2016','2017','2018','2019','2020','2021']
plt.plot(x_axis, TOTALt5['American'], 'r-o', label='American')
plt.plot(x_axis, TOTALt5['Chinese'], 'b--^', label='Chinese')
plt.plot(x_axis, TOTALt5['Italian'], 'g--s', label='Italian')
plt.plot(x_axis, TOTALt5['Japanese'], 'c--*', label='Japanese')
plt.plot(x_axis, TOTALt5['Mexican'], 'm--s', label='Mexican')
plt.show()
```
